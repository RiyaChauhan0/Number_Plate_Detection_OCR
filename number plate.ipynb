```python
import cv2
from google.colab.patches import cv2_imshow
img = cv2.imread('porche.jpg')
number_plate = cv2.CascadeClassifier('/content/haarcascade_russian_plate_number.xml')
demo_plate = number_plate.detectMultiScale(img,1.1,10)
for (x,y,w,h) in demo_plate:
  cv2.rectangle(img,(x,y),(x+w,y+h),(255,255,255),4)
  roi = img[y:y+h,x:x+w] # Extracting required number plate
  cv2_imshow(roi)
print('\n')
cv2_imshow(img)
```
